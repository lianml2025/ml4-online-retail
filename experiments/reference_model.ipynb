```python
##Hasn't been analyzed nor tested - this is a sample initial model from our friend

import pandas as pd
from datetime import timedelta
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load cleaned retail data
df = pd.read_excel("../data/cleaned_retail.xlsx", engine="openpyxl")

# Filter out cancelled transactions
df_clean = df[df['CancellationFlag'].isnull() | (df['CancellationFlag'].astype(str).str.strip() == '')]

# Aggregate customer features
customer_df = df.groupby('CustomerID').agg({
    'Quantity': 'sum',
    'Subtotal': 'sum',
    'InvoiceNo': 'nunique',
    'UnitPrice': 'mean',
    'StockCode': pd.Series.nunique
}).rename(columns={
    'Quantity': 'TotalQuantity',
    'Subtotal': 'TotalSpend',
    'InvoiceNo': 'NumTransactions',
    'UnitPrice': 'AvgUnitPrice',
    'StockCode': 'ProductDiversity'
})

# Normalize features
scaled = StandardScaler().fit_transform(customer_df)

# Cluster
kmeans = KMeans(n_clusters=5, random_state=42)
customer_df['Cluster'] = kmeans.fit_predict(scaled)
```
